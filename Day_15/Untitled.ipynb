```python
def produce_row(grid):
    grid_1 = add_1_to_grid(grid)
    grid_2 = add_1_to_grid(grid_1)
    grid_3 = add_1_to_grid(grid_2)
    grid_4 = add_1_to_grid(grid_3)
    grid_5 = add_1_to_grid(grid_4)
    grid_6 = add_1_to_grid(grid_5)
    grid_7 = add_1_to_grid(grid_6)
    grid_8 = add_1_to_grid(grid_7)
    grid_9 = add_1_to_grid(grid_8)
    row = [grid[i] + grid_1[i] + grid_2[i] + grid_3[i] + grid_4[i] for i in range(len(grid))]
    return row
big_grid = []
big_grid.extend(produce_row(grid))
big_grid.extend(produce_row(add_1_to_grid(grid)))

big_grid.extend(produce_row(add_1_to_grid(add_1_to_grid(grid))))

big_grid.extend(produce_row(add_1_to_grid(add_1_to_grid(add_1_to_grid(grid)))))
big_grid.extend(produce_row(add_1_to_grid(add_1_to_grid(add_1_to_grid(add_1_to_grid(grid))))))
# pprint(big_grid)
```

```python
def add_1_to_grid(grid):
    new_grid = []
    for row in grid:
        new_grid.append("")
        for char in row:
            if char == "9":
                new_grid[-1] += "1"
            else:
                new_grid[-1] += str(int(char)+1)
    return new_grid
```

```python
def node_num(r, c):
    return r * 500 + c
```

```python
import networkx as nx
g = nx.DiGraph()
num_rows = len(grid)
row_length = len(grid[0])
size = num_rows * row_length
print(num_rows, size)
g.add_nodes_from(range(size))
for i in range(size):
    row = i // row_length
    col = i % row_length
    # print(f"({row}, {col})")
    if row > 0:
        g.add_edge(node_num(row-1, col), node_num(row, col), weight=int(grid[row][col]))
    if row < num_rows-1:
        g.add_edge(node_num(row+1, col), node_num(row, col), weight=int(grid[row][col]))
    if col > 0:
        g.add_edge(node_num(row, col-1), node_num(row, col), weight=int(grid[row][col]))
    if col < row_length-1:
        g.add_edge(node_num(row, col+1), node_num(row, col), weight=int(grid[row][col]))
# g.edges()
```

```python
import networkx as nx
def construct_network(grid):
    g = nx.DiGraph()
    num_rows = len(grid)
    row_length = len(grid[0])
    size = num_rows * row_length
    print(num_rows, size)
    g.add_nodes_from(range(size))
    for i in range(size):
        row = i // row_length
        col = i % row_length
        # print(f"({row}, {col})")
        if row > 0:
            g.add_edge(node_num(row-1, col), node_num(row, col), weight=int(grid[row][col]))
        if row < num_rows-1:
            g.add_edge(node_num(row+1, col), node_num(row, col), weight=int(grid[row][col]))
        if col > 0:
            g.add_edge(node_num(row, col-1), node_num(row, col), weight=int(grid[row][col]))
        if col < row_length-1:
            g.add_edge(node_num(row, col+1), node_num(row, col), weight=int(grid[row][col]))
    print(len(g.nodes))
    print(len(g.edges))
    return (g, size)
(graph, size) = construct_network(big_grid)
```

```python
print(nx.number_strongly_connected_components(graph))
print(nx.single_source_dijkstra(graph, 0, size-1))
```
